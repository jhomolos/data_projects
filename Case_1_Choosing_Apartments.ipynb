```python
#import the pandas library
import pandas as pd
```

```python
#import dataset to be analysis
aluguel_casa = pd.read_excel("houses_rent.xlsx")
```

```python
#displaying the first 10 entries in dataset
aluguel_casa.head(10)
```

```python
#display the last 10 entries in dataset
aluguel_casa.tail(10)
```

```python
#renaming columns as a better programming practice
aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces", "rent amount":"rent_amount", "property tax":"property_tax", "fire insurance":"fire_insurance"})
```

```python
#printing dataset after renaming columns
aluguel_casa.head()
```

```python
#visualizing the data types of the columns
aluguel_casa.dtypes
```

```python
#redefining columns types
aluguel_casa = aluguel_casa.astype({"rent_amount":"float", "property_tax":"float", "fire_insurance":"float", "total":"float"})
aluguel_casa.dtypes
```

```python
#printing dataset
aluguel_casa.head()
```

```python
#printing relevant information concerning the dataset
aluguel_casa.info()
```

```python
#printing relevant metrics concerning the data
aluguel_casa[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe()
```

```python
#filtering apartments with rent costing less than USD 4,000.
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
#filtering apartments where animals are accepted and rent costing less than USD 4,000
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")]
aluguel_casa.head()
```

```python
#using the query method to apply the same filters
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
#filtering 2 or 3 room apartments, whose rent costs less than USD 4,000 and animals are accepted
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))]
aluguel_casa.head()
```

```python
#using the query method to apply the same filters
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa
```

```python
aluguel_casa.describe()
```

```python
#creating a variable agg_city_price to group by city and to display the average price of rents
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
#creates the variable agg_bath_room_price by grouping by bathroom and room amounts, as well their average price of rents
agg_bath_room_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```

```python
#imports the seaborn library
import seaborn as sns
```

```python
#plots the quantity of rooms x rent price using the seaborn library
ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color = "blue");
ax.set(xlabel = "qtd quartos", ylabel = "media do preo");
```

```python
#plots a "quantity of bathrooms x rent price" graph using the seaborn library
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color = "green");
ax.set(xlabel = "qtd banheiros", ylabel = "media do preo");
```

```python
#it turns out the Porto has the best cost-benefit, and therefore, we filter our data to display only the apartments that consider the previous filter and city = "Porto"
aluguel_casa = aluguel_casa.query("city == 'Porto'")
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
#it groups all the apartments that fulfill the renter's preferences and groups the apartments by the floor where they are located
aluguel_casa.groupby("floor").index.nunique().sort_values(ascending=False)
```

```python
#it calculates the average price of apartments according to the floor they are located
agg_floor_price = aluguel_casa.groupby("floor")["total"].agg("mean").reset_index()
agg_floor_price
```

```python
#plots a bar gralph to display the average price of rents according to the floor where the apartments are located
sns.barplot(data = agg_floor_price, x="floor", y="total", color = "blue")
```

```python
#it turns out that the sixteenth floor is the floor with the best cost-benefit relation. Our premise is that the apartments located at higher floors have more value.
sixteenth_floor = aluguel_casa.query("floor == 16")
sixteenth_floor.head()
```
